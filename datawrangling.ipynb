```python
image_prediction_df = pd.read_csv('image-predictions.tsv', sep='\t')
```

```python
consumer_key = ''
consumer_secret = ''
access_token = ''
access_secret = ''

auth = OAuthHandler(consumer_key, consumer_secret)
auth.set_access_token(access_token, access_secret)

#api = tweepy.Client(auth, wait_on_rate_limit=True)

bearer_token=''
# create instance
client = tweepy.Client(bearer_token=bearer_token, wait_on_rate_limit=True)
```

```python
tweetlist = []
notweetlist = []
count = 0


tweet_ids = image_prediction_df.tweet_id.values


# Query Twitter's API for JSON data for each tweet ID in the Twitter archive
count = 0
fields=["id","public_metrics","source","text"]

for tweet_id in image_prediction_df['tweet_id']:
    temp_dict={}
    try:
        print(f'{str(count)}: {str(tweet_id)}')
        count= count + 1
        temp=client.get_tweet(tweet_id, tweet_fields=fields)
        temp_dict['tweet_id']=temp.data.id
        temp_dict['text']=temp.data.text
        temp_dict['source']=temp.data.source
        temp_dict['retweet_count']=temp.data.public_metrics['retweet_count']
        temp_dict['reply_count']=temp.data.public_metrics['reply_count']
        temp_dict['like_count']=temp.data.public_metrics['like_count']
        temp_dict['quote_count']=temp.data.public_metrics['quote_count']
        tweetlist.append(temp_dict)
        print('Pass')
    except Exception as e:
        notweetlist.append(tweet_id)
        print('Fail')
```
